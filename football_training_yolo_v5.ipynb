```python
from clearml import Dataset
dataset = Dataset.get(
        dataset_id='e147786bec9a4896a92e8da77d23a8cc',  
        only_completed=True, 
        only_published=False, 
        alias='my_third_training_data_image640'
)

dataset_path = dataset.get_local_copy()
```

```python
data_yaml_path = dataset_path + "/data.yaml"
```

```python
data_yaml_path
```

```python
from ultralytics import YOLO
from clearml import Task

# Step 1: Creating a ClearML Task
task = Task.init(project_name="football-analysis", task_name="fourth_train_img640")

# Step 2: Selecting the YOLOv8 Model
model_variant = "yolov5xu"
task.set_parameter("model_variant", model_variant)

# Step 3: Loading the YOLOv8 Model
model = YOLO(f"{model_variant}.pt")

# Step 4: Setting Up Training Arguments
args = dict(data=data_yaml_path, epochs=100, imgsz = 640)
task.connect(args)
task.connect(dataset, name='input_dataset') 

# Step 5: Initiating Model Training
results = model.train(**args)

task.close()
```
